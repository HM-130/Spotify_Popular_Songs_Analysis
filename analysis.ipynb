```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate average values for each year
avg_1970 = df_1970[['danceability', 'energy', 'valence']].mean()
avg_2024 = df_2024[['danceability', 'energy', 'valence']].mean()

# Combine averages into a DataFrame for plotting
avg_df = pd.DataFrame({'1970': avg_1970, '2024': avg_2024})

# Plotting
avg_df.plot(kind='bar', figsize=(12, 6), color=['blue', 'orange'])
plt.title('Comparison of Average Song Characteristics: 1970s vs 2024')
plt.ylabel('Average Value')
plt.xlabel('Characteristics')
plt.legend(['1970', '2024'])
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate average loudness for each year
avg_loudness_1970 = df_1970['loudness'].mean()
avg_loudness_2024 = df_2024['loudness'].mean()

# Create a DataFrame with these averages
loudness_df = pd.DataFrame({
    'Year': ['1970', '2024'],
    'Average Loudness': [avg_loudness_1970, avg_loudness_2024]
})

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(loudness_df['Year'], loudness_df['Average Loudness'], color=['blue', 'orange'])
plt.title('Average Loudness of Songs: 1970 vs. 2024')
plt.ylabel('Loudness (dB)')
plt.xlabel('Year')
plt.show()
```
